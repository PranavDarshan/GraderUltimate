```python
import os
import requests

pdf_path_2 = "operating_systems_book_2.pdf"


if not os.path.exists(pdf_path_2):
  print(f"[INFO] File doesn't exist downloading!")

  url = "https://os.ecci.ucr.ac.cr/slides/Abraham-Silberschatz-Operating-System-Concepts-10th-2018.pdf"

  #local file name to save the downloaded file
  filename = pdf_path_2

  response = requests.get(url)

  if response.status_code == 200:
    #open the file and save it

    with open(filename, "wb") as file:
      file.write(response.content)

    print(f"[INFO] the file has been downloaded and saved as {filename}")
  else:
    print(f"[INFO] failed to download the file. Status code {response.status_code}")

else:
  print(f"File {pdf_path_2} exists.")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.document_loaders import WebBaseLoader
from langchain_community.vectorstores import FAISS
from langchain_core.vectorstores import VectorStoreRetriever
```

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader(pdf_path_3)
docs = loader.load()
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 5000,
                                               chunk_overlap = 1000)
documents = text_splitter.split_documents(docs)
```

```python
import fitz
from tqdm.auto import tqdm
def text_formatter(text: str) -> str:
  """Performs minor formatting on text"""
  cleaned_text = text.replace("\n", " ").strip()
  ## potentially other cleaning can take place here

  return cleaned_text

for doc in tqdm(documents):
  doc.page_content = text_formatter(doc.page_content)

```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

model_name = "sentence-transformers/all-MiniLM-L6-v2"
model_kwargs = {'device': 'cuda:0'}
encode_kwargs = {'normalize_embeddings': False}
hf = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
library = FAISS.from_documents(documents, hf)
```

```python
Query1 = "what is multithreading?"
Query_answer = library.similarity_search(Query1)
```

```python
print(Query_answer[0].page_content)
```
